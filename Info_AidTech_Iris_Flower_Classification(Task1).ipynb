```python
X=iris_data.drop('species', axis=1)
y=iris_data['species']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42, test_size=0.25)
```

# **Decision Tree Classfier**

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)

print("Score of Test data is:",dt.score(X_test,y_test))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

fig, axis = plt.subplots(figsize=(10,8))
tree.plot_tree(dt, feature_names=X.columns, class_names=y.unique(), filled=True, ax=axis)
plt.show()
```

```python
dt.predict(X_test)
```

```python
# Lets shuffle values in columns and try, this become new data
from random import shuffle

sepal_length = list(iris_data["sepal_length"])
shuffle(sepal_length)

sepal_width = list(iris_data["sepal_width"])
shuffle(sepal_width)

petal_length = list(iris_data["petal_length"])
shuffle(petal_length)

petal_width = list(iris_data["petal_width"])
shuffle(petal_width)

new_iris_data = pd.DataFrame({"sepal_length":sepal_length, "sepal_width":sepal_width, "petal_length":petal_length, "petal_width":petal_width})
new_iris_data.head()
```

```python
dt.predict(new_iris_data)
```

```python
def classify_iris_type(s_length,	s_width,	p_length,	p_width):
  test = {"sepal_length":[s_length],	"sepal_width":[s_width],	"petal_length":[p_length],	"petal_width":[p_width]}
  pred = dt.predict(pd.DataFrame(test))
  print("Iris type is:", pred[0])
```

```python
go=1
while(go==1):
  sl=float(input("Enter sepal_length: "))
  sw=float(input("Enter sepal_width: "))
  pl=float(input("Enter petal_length: "))
  pw=float(input("Enter petal_width: "))
  classify_iris_type(sl, sw, pl, pw)
  go=int(input("Want to cont ?(0/1): "))
```

# **KNN**

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1, metric='minkowski', p=2 )
knn.fit(X_train, y_train)
print("Score using KNN is:",knn.score(X_test,y_test))
```
